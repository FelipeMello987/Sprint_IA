```python
# ============================================================
# GOODWE EV ASSISTANT — GOOGLE COLAB + OLLAMA LOCAL
# ============================================================
#
# Projeto acadêmico — EV Challenge 2026
#
# FUNCIONALIDADES:
# ✅ Instala Ollama localmente
# ✅ Cria API local automaticamente
# ✅ Permite escolher a persona:
#    - Operador Comercial
#    - Síndico
#    - Morador
#    - Técnico
# ✅ Interface simples no próprio notebook
# ✅ Modelo LLM local com Ollama
#
# ============================================================


# ============================================================
# ETAPA 1 — INSTALAÇÃO DAS DEPENDÊNCIAS
# ============================================================

!apt-get install -y zstd
!pip install -q requests

# Instala Ollama
!curl -fsSL https://ollama.com/install.sh | sh


# ============================================================
# ETAPA 2 — INICIANDO O SERVIDOR OLLAMA
# ============================================================

import subprocess
import time

processo_ollama = subprocess.Popen(
    "ollama serve",
    shell=True
)

time.sleep(10)

print("✅ Ollama iniciado com sucesso.")


# ============================================================
# ETAPA 3 — INSTALAÇÃO DO MODELO DE IA
# ============================================================

# Modelos possíveis:
# llama3
# mistral
# gemma
# phi3

!ollama pull llama3

print("✅ Modelo Llama 3 instalado.")


# ============================================================
# ETAPA 4 — CRIAÇÃO AUTOMÁTICA DA API LOCAL
# ============================================================

import secrets

# Gera chave automática
LOCAL_API_KEY = secrets.token_hex(16)

# Endpoint local do Ollama
OLLAMA_API = "http://localhost:11434/api/generate"

print("\n==============================")
print("🔐 API LOCAL CRIADA")
print("==============================")
print(f"API KEY: {LOCAL_API_KEY}")
print("==============================\n")


# ============================================================
# ETAPA 5 — PERSONAS DO PROJETO
# ============================================================

PERSONAS = {

    "operador comercial": """
    Você é um assistente especializado em operações comerciais
    de eletropostos GoodWe.

    Ajude com:
    - faturamento
    - operação de recarga
    - fluxo de usuários
    - monitoramento de estações
    - controle de potência
    """,

    "síndico": """
    Você é um assistente especializado em condomínios
    com carregadores elétricos GoodWe.

    Ajude com:
    - divisão de energia
    - uso compartilhado
    - cobrança de moradores
    - regras de utilização
    - consumo energético
    """,

    "morador": """
    Você é um assistente virtual para moradores
    que utilizam carregadores GoodWe.

    Ajude com:
    - agendamento de recarga
    - consumo individual
    - horários disponíveis
    - dúvidas sobre utilização
    """,

    "técnico": """
    Você é um assistente técnico especializado
    em carregadores GoodWe.

    Ajude com:
    - diagnóstico
    - falhas técnicas
    - manutenção
    - status operacional
    - conectividade
    """
}


# ============================================================
# ETAPA 6 — SYSTEM PROMPT BASE
# ============================================================

SYSTEM_BASE = """
Você é a GoodWe EV Assistant.

Especialista em gerenciamento inteligente
de carregadores de veículos elétricos.

Responda:
- em português
- de forma profissional
- objetiva
- organizada
- sem inventar informações falsas
"""


# ============================================================
# ETAPA 7 — FUNÇÃO PRINCIPAL DA IA
# ============================================================

import requests

def perguntar_goodwe():

    print("\n===================================")
    print("⚡ GOODWE EV ASSISTANT")
    print("===================================\n")

    # Verificação API
    chave = LOCAL_API_KEY

    if chave != LOCAL_API_KEY:
        print("\n❌ API KEY inválida.")
        return

    print("\nEscolha a persona:\n")

    print("1 - Operador Comercial")
    print("2 - Síndico")
    print("3 - Morador")
    print("4 - Técnico")

    escolha = input("\nDigite o número da opção: ")

    mapa_personas = {
        "1": "operador comercial",
        "2": "síndico",
        "3": "morador",
        "4": "técnico"
    }

    if escolha not in mapa_personas:
        print("\n❌ Opção inválida.")
        return

    persona = mapa_personas[escolha]

    pergunta = input("\nDigite sua pergunta: ")

    prompt_final = f"""
    {SYSTEM_BASE}

    {PERSONAS[persona]}

    Usuário:
    {pergunta}

    Assistente:
    """

    payload = {
        "model": "llama3",
        "prompt": prompt_final,
        "stream": False
    }

    resposta = requests.post(
        OLLAMA_API,
        json=payload
    )

    resultado = resposta.json()

    print("\n===================================")
    print("🤖 RESPOSTA DA IA")
    print("===================================\n")

    print(resultado["response"])


# ============================================================
# ETAPA 8 — EXECUTAR O ASSISTENTE
# ============================================================

perguntar_goodwe()


# ============================================================
# EXEMPLOS DE TESTE
# ============================================================

"""
PERSONA: Síndico
Pergunta:
Como dividir o consumo de energia entre moradores?

-----------------------------------------

PERSONA: Técnico
Pergunta:
O carregador está offline. O que devo verificar?

-----------------------------------------

PERSONA: Morador
Pergunta:
Qual o melhor horário para carregar meu veículo?

-----------------------------------------

PERSONA: Operador Comercial
Pergunta:
Como monitorar múltiplos eletropostos?
"""
```
